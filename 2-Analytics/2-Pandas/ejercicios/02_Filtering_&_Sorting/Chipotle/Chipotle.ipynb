### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
```

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv(url, sep='\t')
```

```python
chipo.info()
```

# Name of the max valued product

```python
chipo['item_price'] = list(map(lambda x: float(x.replace('$', '')), chipo['item_price']))

```

```python
# chipo['item_price'] = chipo['item_price'].str.replace('$', '').astype(float)

chipo['item_price'].info()
```

```python
chipo['item_price'].max()
```

```python
chipo[['item_price', 'item_name']].sort_values(by='item_price', ascending=False).head(1)
```

### Step 4. How many products cost more than $10.00?

```python
mask = chipo.item_price > 10

chipo.item_price[mask]

```

```python
chipo.item_price[mask].count()
```
